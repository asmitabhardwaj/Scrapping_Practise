# IMPORT MODULES

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

# GET Page Content from url

```python
url="https://zeenews.india.com/"
page= requests.get(url)
page
```

```python
page.content
```

```python
Soup=bs(page.content,"html.parser")
print(Soup.prettify())
```

```python
#find image src link
image_tag=Soup.find('div',{'class':"top-story-desktop big-news-thumb bignews-design"})
image_tag
```

```python
anchors=image_tag.find('a')
anchors
```

```python
news_link=image_tag.find('a')['href']
news_link
```

```python
news_title=image_tag.find('a')['title']
news_title
```

```python
image_url=anchors.find('img')['src']
image_url
```

# To download the image

```python
image= requests.get(image_url)
image
```

```python
#store image in file
with open('image.jpg','wb') as file:
    for chunk in image.iter_content(chunk_size=1024):
        file.write(chunk)


```
